# Exploratory data analysis and Data visualization


1.   Exploratory Analysis
2.   Data Visualization

```python
#import modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt # for plotting graphs
import seaborn as sns # for plotting graphs
%matplotlib inline
```

```python
import matplotlib.pyplot as plt

# Plot histogram untuk kolom 'loan_amnt'
plt.hist(data['loan_amnt'], bins=30, edgecolor='black')
plt.title('Histogram of Loan Amount')
plt.xlabel('Loan Amount')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Plot scatter plot antara 'loan_amnt' dan 'int_rate'
plt.scatter(data['loan_amnt'], data['int_rate'], alpha=0.5)
plt.title('Scatter Plot: Loan Amount vs Interest Rate')
plt.xlabel('Loan Amount')
plt.ylabel('Interest Rate')
plt.grid(True)
plt.show()
```

```python
# Plot box plot untuk kolom 'annual_inc'
plt.boxplot(data['annual_inc'])
plt.title('Box Plot of Annual Income')
plt.ylabel('Annual Income')
plt.grid(True)
plt.show()
```

```python
# Plot bar plot untuk kolom 'purpose'
purpose_counts = data['purpose'].value_counts()
plt.barh(purpose_counts.index, purpose_counts.values)
plt.title('Bar Plot of Loan Purposes')
plt.xlabel('Count')
plt.ylabel('Purpose')
plt.grid(axis='x')
plt.show()
```

```python
# Plot pie chart untuk kolom 'grade'
grade_counts = data['grade'].value_counts()
plt.pie(grade_counts, labels=grade_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart of Loan Grades')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Plot line plot untuk kolom 'revol_bal'
plt.plot(data['revol_bal'])
plt.title('Line Plot of Total Revolving Balance')
plt.xlabel('Index')
plt.ylabel('Revolving Balance')
plt.grid(True)
plt.show()
```

```python
# Plot bar plot untuk kolom 'home_ownership'
home_counts = data['home_ownership'].value_counts()
plt.bar(home_counts.index, home_counts.values)
plt.title('Bar Plot of Home Ownership')
plt.xlabel('Home Ownership')
plt.ylabel('Count')
plt.grid(axis='y')
plt.show()
```

```python
# Plot scatter plot antara 'total_acc' dan 'dti'
plt.scatter(data['total_acc'], data['dti'], alpha=0.5)
plt.title('Scatter Plot: Total Accounts vs Debt-to-Income Ratio')
plt.xlabel('Total Accounts')
plt.ylabel('Debt-to-Income Ratio')
plt.grid(True)
plt.show()
```
